```python
# Scatter
from matplotlib import pyplot as plt
x1 = [1,2,3,4,5]
y1 = [0,1,5,2,5]
x2 = [1,2,3,4,5]
y2 = [0,2,5,2,1]
color = ["hotpink","yellow","cyan","magenta","green"]
# First One
plt.subplot(1,2,1)
plt.scatter(x1,y1,c="cyan")
plt.title("First")


# Second One
plt.subplot(1,2,2)
plt.scatter(x2,y2,c=color)
plt.title("Second")


# Final
plt.suptitle("Data Sets")
plt.show()
```

```python
# Scatter
from matplotlib import pyplot as plt
from random import uniform as U

x1 = [1,2,3,4,5]
y1 = [0,1,5,2,5]
x2 = [1,2,3,4,5]
y2 = [0,2,5,2,1]
color = ["hotpink","yellow","cyan","magenta","green"]
sizes = [100,200,300,400,500]
a = U(0,1)
# First One
plt.subplot(1,2,1)
plt.scatter(x1,y1,c="cyan",s = sizes,alpha = a)
plt.title("First")


# Second One
plt.subplot(1,2,2)
plt.scatter(x2,y2,c=color,alpha = a)
plt.title("Second")


# Final
plt.suptitle("Data Sets")
plt.show()
```

```python
# Bars
from matplotlib import pyplot as plt
x = ["A","B","C","D","E"]
y = [1,2,3,4,2]
col = ["magenta","yellow","green","red","cyan"]
plt.bar(x,y,color = col,width = 0.5)
plt.title("Bar",loc="left")
plt.xlabel("Letters")
plt.ylabel("Numbers")
plt.show()
```

```python
# Horizontal Bars
from matplotlib import pyplot as plt
x = ["A","B","C","D","E"]
y = [1,2,3,4,2]
col = ["magenta","yellow","green","red","cyan"]
plt.barh(x,y,color = col,height=0.5)
plt.title("Bar",loc="left")
plt.xlabel("Letters")
plt.ylabel("Numbers")
plt.show()
```

```python
# Pie
from matplotlib import pyplot as plt
y = [20,85,90,23,10]
mylabels = ["Apple","Banana","Grapes","Weed","Tobacco"]
myexplode = [0,0.2,0.4,0,0]
mycolors = ["black","blue","cyan","magenta","green"]
plt.pie(y,labels = mylabels,explode=myexplode,shadow = True,colors = mycolors)
plt.legend(title="Fruits")
plt.show()
```
